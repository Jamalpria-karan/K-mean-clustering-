```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv(r'Downloads/marketing_campaign.csv',sep='\t')
df
```

```python
df['age']=2021-df['Year_Birth']
df.head()
```

```python
df['Education'].value_counts()
```

```python
edu_nums = {"Education":     {"Basic": 1, "2n Cycle": 2,"Graduation":3,"Master":4,"PhD":5}}
df = df.replace(edu_nums)
df.head()
```

```python
df['Marital_Status'].value_counts()
```

```python
df.shape
```

```python
a=['Absurd','YOLO']
df = df[~df['Marital_Status'].isin(a)]
df.shape
```

```python
mar_nums = {"Marital_Status":     {"Single": 1,"Alone":1,"Married": 2,"Together":2,"Divorced":3,"Widow":4}}
df = df.replace(mar_nums)
df.head()
```

```python
df.drop(['ID','Year_Birth','Teenhome','Dt_Customer','Complain','Z_CostContact','Z_Revenue','Response'], axis = 1,inplace=True)
df.head()
```

```python
df.info()
```

```python
df['num_total_purchase']=df['NumDealsPurchases']+df['NumStorePurchases']+df['NumCatalogPurchases']+df['NumWebPurchases']
df.info()
```

```python
df.drop(['NumStorePurchases','NumCatalogPurchases','NumWebPurchases','NumDealsPurchases'], axis = 1,inplace=True)
df.info()
```

```python
df['Income'].unique()
```

```python
df['Income'].fillna(df['Income'].median())
```

```python
df['offer_accepted']=df['AcceptedCmp1']+df['AcceptedCmp2']+df['AcceptedCmp3']+df['AcceptedCmp4']+df['AcceptedCmp5']
df.head()
```

```python
df.drop(['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5'], axis = 1,inplace=True)
df.info()
```

```python
df['offer_accepted'].unique()
```

```python
df.loc[df['offer_accepted'] >0, 'offer_accepted'] = 1
```

```python
df['Education'].unique()
```

```python
df['offer_accepted'].unique()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2)
```

```python
train=train.dropna()
```

```python
X=np.array(train.drop(['offer_accepted'], 1).astype(float))
```

```python
y=np.array(train['offer_accepted'])
```

```python
train.info()
```

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=300,
    n_clusters=2, n_init=10, n_jobs=1, precompute_distances='auto',
    random_state=None, tol=0.0001, verbose=0)
```

```python
correct = 0
for i in range(len(X)):
    predict_me = np.array(X[i].astype(float))
    predict_me = predict_me.reshape(-1, len(predict_me))
    prediction = kmeans.predict(predict_me)
    if prediction[0] == y[i]:
        correct += 1
```

```python
print(correct/len(X))
```

```python
kmeans = kmeans = KMeans(n_clusters=2, max_iter=600, algorithm = 'auto')
kmeans.fit(X)
```
